# 2.0 Harmonic Excitation
Reference: Dynamics of Structures, Clough and Penzien

We will now consider the response of SDOF systems to a harmonic force. A harmonic force is one whose magnitude varies sinusoidally, e.g.

\begin{equation}
p(t) =P_0\sin(\omega t) 
\end{equation}

This sinusoidal forcing function has an amplitude $P_0$ with angular frequency $\omega$. Understanding the behaviour of SDOF systems to harmonic loading allows us to establish our understanding of a number of important concepts, such as **dynamic magnification and resonance** which we'll discuss further below.

The other reason we focus on harmonic loading initially is because it will provide us with the tools to analyse much more complex loading further down the road using a technique known as Fourier decomposition.

While Fourier decomposition is outside the scope of this module, the basic idea is that any periodic force (regularly repeating but not harmonic) may be decomposed into a series of harmonic forces. This series consists of a fundamental harmonic and a succession of higher order harmonics, all of which can be represented by sine/cosine functions of varying phase, amplitude and increasing frequencies. Thus, the analysis of a dynamic system subject to periodic loading can be reduced to multiple analyses of the system subject to harmonic loading functions. 

## 2.1 Response to harmonic excitation

Consider the SDOF spring-mass-damper system you're familiar with, Fig. 1. However, now it is subject to a harmonic forcing function.

![title](img/img1.jpg)
<center>Fig 1. SDOF system subject to harmonic excitation.</center>

Building on what you know from the study of free vibration, the equation of motion that describes this forced system is,

\begin{equation}
m\ddot{U} + c\dot{U} + kU=P_0\sin(\omega t)
\end{equation}

Our objective now is to solve this differential equation to determine an expression for the displacement as a function of time, $U(t)$. Remember that this is a non-homogeneous second order differential equation and we have seen previously that the solution to such an equation has two parts:

1. The complementary solution (also the solution to the equivalent homogeneous equation)


2. The particular solution

We've already developed the complementary solution, this is the free vibration response. So what remains now is to find the particular solution. This proceeds in the usual way with the term on the right hand side of the equal sign suggesting the form of a trial solution (refer back to how we solved the buckling equations). 

Since the right hand side of this equation is a sinusoid, the trial solution will be a generalised sinusoid,

\begin{equation}
U(t) = A\sin(\omega t) + B\cos(\omega t)
\end{equation}

We now differentiate this trial solution so it can be substituting back into our original differential equation,

\begin{equation}
\dot{U}(t) = \omega A\cos(\omega t) -\omega B\sin(\omega t)
\end{equation}

\begin{equation}
\ddot{U}(t) = -\omega^2 A\sin(\omega t) -\omega^2 B\cos(\omega t)
\end{equation}

Substituting these expressions back into our differential equation yields,

\begin{equation}
m\left[-\omega^2 A\sin(\omega t) -\omega^2 B\cos(\omega t) \right] + c\left[\omega A\cos(\omega t) -\omega B\sin(\omega t) \right] + k\left[A\sin(\omega t) + B\cos(\omega t) \right] = P_0\sin(\omega t)
\end{equation}

Rearranging yields,

\begin{equation}
(k-m\omega^2)\left[A\sin(\omega t) + B\cos(\omega t)\right] + c\omega\left[A\cos(\omega t)-B\sin(\omega t)\right] = P_0\sin(\omega t)
\end{equation}

We can obtain expressions for the constants $A$ and $B$ by collecting sine and cosine terms together. Collecting sine terms yields (after sine terms cancelled),

\begin{equation}
A(k-m\omega^2) - B(c\omega) = P_0
\end{equation}

Collecting and cancelling out cosine terms yields,

\begin{equation}
A(c\omega) + B(k-m\omega^2) =0
\end{equation}

These two simultaneous equations are used to obtain the following expressions for $A$ and $B$,

\begin{equation}
A = \frac{k-m\omega^2}{(k-m\omega^2)^2 + (c\omega)^2} P_0
\end{equation}

\begin{equation}
B = \frac{-c\omega}{(k-m\omega^2)^2 + (c\omega)^2}P_0
\end{equation}

Finally, these expressions can be substituted back into our assumed solution for $U(t)$, to yield the following expression for the particular solution,

\begin{equation}
\boxed{
U_p(t) = \frac{P_0}{(k-m\omega^2)^2 + (c\omega)^2}\left[(k-m\omega^2)\sin(\omega t) - c\omega \cos(\omega t) \right]
}
\end{equation}

Combining the complimentary and particular solutions we obtain the complete expression for the response of the SDOF system under harmonic excitation,

\begin{equation}
U(t) = U_c(t) + U_p(t)
\end{equation}

\begin{equation}
\boxed{
U(t) = e^{-\xi \omega_nt}\:\left[A\sin(\omega_d t) +B\cos(\omega_d t)\right] + \frac{P_0}{(k-m\omega^2)^2 + (c\omega)^2}\left[(k-m\omega^2)\sin(\omega t)-c\omega\cos(\omega t)\right]
}
\end{equation}

Notice that the complimentary solution describes oscillations at the damped natural frequency of the structure $\omega_d$, while the particular solution describes oscillations at the forcing frequency. 

As we've said, the complimentary solution describes the free vibration response. This corresponds to transient behaviour that occurs at the beginning of the oscillatory behaviour when the force is first applied. This transient response decays away exponentially leaving only the response described by the particular solution. 

The particular solution is therefore often referred to as the steady-state solution. For a dynamic system experiencing harmonic forcing, the steady state solution is of more interest than the transient solution.

Note also that in the equation above, the transient portion of the response still contains unknown constants of integration A and B. These can be determined from knowledge of the system initial conditions (position and velocity). But as we've said, this portion of the response decays to zero and as such will not be of interest when considering the steady state oscillating behaviour. 

### 2.1.1 Dimensionless response equation

The equation above that describes the complete response is often presented in an alternative form that makes use of dimensionless ratios. If we define $\beta$ as the frequency ratio,

\begin{equation}
\beta = \frac{\omega}{\omega_n}
\end{equation}

where $\omega$ is the forcing frequency, we can restate the response equation as,

\begin{equation}
u(t) = e^{-\xi \omega_nt}\:\left[A\sin(\omega_d t) +B\cos(\omega_d t)\right] + \frac{P_0}{k}\left[\frac{1}{(1-\beta^2)^2+(2\xi\beta)^2}\right]\left[(1-\beta^2)\sin(\omega t)-2\xi\beta\cos(\omega t)\right]
\end{equation}

Focusing now on the steady-state response only, we have,

\begin{equation}
U_p(t) = \frac{P_0}{k}\left[\frac{1}{(1-\beta^2)^2+(2\xi\beta)^2}\right]\left[(1-\beta^2)\sin(\omega t)-2\xi\beta\cos(\omega t)\right]
\end{equation}

We notice that the response has two components, one in phase with the forcing function, 

\begin{equation}
\frac{P_0}{k}\left[\frac{1}{(1-\beta^2)^2+(2\xi\beta)^2}\right]\:\times \:(1-\beta^2)\sin(\omega t)
\end{equation}

and one $90^\circ$, out of phase with it (indicated by the use of the cosine term),

\begin{equation}
\frac{P_0}{k}\left[\frac{1}{(1-\beta^2)^2+(2\xi\beta)^2}\right]\:\times \:-2\xi\beta\cos(\omega t)
\end{equation}

The out of phase component is introduced by the fact that there is damping in the system. We can think of this damping as (phase) shifting the response out of phase with the applied force causing the response to 'lag' behind the force. This 'lag' is seen as a phase difference between applied force and response.

## 2.2 Steady-state response

A helpful way to think about the steady-state response is in terms of its relationship to the forcing function that is generating it. To get started we can represent the two components of the steady-state response as rotating vectors, but before we sketch this out, we'll make things easier to read by making the following temporary substitution,

\begin{equation}
\Omega = \frac{P_0}{k}\left[\frac{1}{(1-\beta^2)^2+(2\xi\beta)^2}\right]
\end{equation}

Now, if we visualise both components of the steady state response as two vectors, $90^\circ$ out of phase, with amplitudes given as indicated in Fig. 2,

![title](img/img2.jpg)
<center>Fig 2. Two components of the steady-state response represented as rotating vectors.</center>

The resulting vector $R$ would have magnitude,

\begin{equation}
|u(t)|=\Omega \sqrt{(1-\beta^2)^2 + (2\xi\beta)^2}
\end{equation}

Reversing the temporary substitution we just made, 

\begin{equation}
|u(t)|=\frac{P_0}{k}\frac{1}{\sqrt{(1-\beta^2)^2 + (2\xi\beta)^2}}
\end{equation}

We can also state that the phase of the resulting steady-state response with reference to the component of resposne in phase with the force is,

\begin{equation}
\phi = \tan^{-1}\frac{2\xi\beta}{1-\beta^2}
\end{equation}

In other words this is the phase the steady-state response makes with the forcing function or the degree to which the steady-state response of the system is out of phase with the force that's generating it. 

### 2.2.1 Dynamic Magnification Factor (DMF)

The dynamic magnification factor is the ratio of the amplitude of steady-state dynamic response to the response due to a static load of the same magnitude. This essentially allows us to easily identify the degree to which the dynamic nature of the force amplifies the structural response. We can state the DMF as, 

\begin{equation}
DMF = \frac{|u(t)|}{\frac{P_0}{k}} = \frac{1}{\sqrt{(1-\beta^2)^2 + (2\xi\beta)^2}}
\end{equation}

The DMF and phase of the response are functions of the forcing frequency, the system's natural frequency and the level of damping in the system. We can get a better sense for the variation by graphing these quantities. 

```python
beta = np.linspace(0,2,500) #Define a range of frequency ratios
Xi = np.array([0.05, 0.1, 0.2, 0.5]) #Define a range of damping ratios

#Plot DMF
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,1.5,1]) 

#Loop through each damping ratio and plot the DMF as a function of beta
for xi in Xi:
    DMF = 1/np.sqrt((1-beta**2)**2 + (2*xi*beta)**2)   
    Label = str(xi*100) + '% damping' 
    axes.plot(beta,DMF,'-',label=Label)

axes.set_xlim([0,2])
axes.set_xlabel('Frequency Ratio, $\omega/\omega_n$')
axes.set_ylabel('DMF')
axes.set_title('Dynamic Magnification Factor')
axes.grid()
axes.legend(loc='upper right')
plt.show()

#Plot Phase Angle
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,1.5,1]) 

#Loop through each damping ratio and plot the phase angle as a function of beta
for xi in Xi:
    angle = np.arctan((2*xi*beta)/(1-beta**2)) #Angle in radians   
    angleDeg = angle*180/math.pi #Angle in degrees     
    angleDeg[250:500] = angleDeg[250:500]+180 #Shift by 180 degrees for beta>1        
    Label = str(xi*100) + '% damping' 
    axes.plot(beta,angleDeg,'-',label=Label)

axes.set_xlim([0,2])
axes.set_ylim([0,180])
axes.set_xlabel('Frequency Ratio, $\omega/\omega_n$')
axes.set_ylabel('Phase Angle in Degrees')
axes.set_title('Phase angle between force and response')
axes.grid()
axes.legend(loc='lower right')
plt.show()
```

Note that the maximum DMF does not occur when $\beta = 1$. The peak dynamic response actually occurs when, 

\begin{equation}
\beta = \sqrt{1-2\xi^2}
\end{equation}

The peak dynamic response in a damped system occurs when the forcing frequency is slightly below the undamped natural frequency. Practically however for small levels of damping, the offset is negligibly small. 

### 2.2.2 Resonance

As the excitation frequency approaches the natural frequency of the system, the steady state amplitude reaches a maximum. This condition is known as resonance and can cause a dynamic response far in excess of the equivalent static force response. 

In civil engineering structures a resonance condition must be avoided if possible. We can see the effects of resonance on bridges that have natural frequencies that coincide with the typical walking or pacing frequency of humans. 

In these relatively rare situations, large bridge oscillations can develop, leading to feedback between the occupying crowd and bridge response. The classic example of this type of positive feedback loop is the London Millennium Bridge which underwent large amplitude lateral oscillations on its opening day in 2000.
